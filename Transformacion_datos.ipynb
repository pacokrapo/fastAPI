```python
import pandas as pd
```

```python
#Cargamos nuestra base de datos en un DataFrame de Pandas:

MoviesDataset = pd.read_csv("DataSets/movies_dataset.csv", sep=',')
```

```python
#Para ver cuántos datos nulos tenemos en nuestro dataset:

MoviesDataset.isnull().sum()
```

Revenue, budget y return

```python
#Para rellenar los valores nulos de la columna revenue:

MoviesDataset['revenue'].fillna(0, inplace=True)
```

```python
#Eliminamos las filas que tengan valor nulo en la columna release_date
MoviesDataset.dropna(subset=['release_date'], inplace=True)

#Creamos la columna release year:
MoviesDataset["release_year"]=None

#Para rellenar los valores de la columna release year:
for i in range(0,len(MoviesDataset)):
    MoviesDataset.release_year.iloc[i] = int(MoviesDataset.release_date.iloc[i][:4])
```

```python
#Creamos la columna Return:
MoviesDataset["return"]=None

#Para transformar los valores de budget a float y transformar los valores que no sean numéricos a 0:

for i in range(0,len(MoviesDataset)):
    try:
        MoviesDataset.budget.iloc[i]=float(MoviesDataset.budget.iloc[i])
    except ValueError:
        MoviesDataset.budget.iloc[i]=0

#Para cargar los datos a la columna revenue:
for i in range(0,len(MoviesDataset)):
    if MoviesDataset.budget.iloc[i] == 0:
        MoviesDataset["return"].iloc[i] = 0
    else:
        MoviesDataset["return"].iloc[i] = (MoviesDataset.revenue.iloc[i])/(MoviesDataset.budget.iloc[i])

```

Columnas innecesarias

```python
#Para eliminar las columnas innecesarias:

columnas_a_eliminar = ["video","imdb_id","adult","poster_path", "homepage", "original_title"]

MoviesDataset.drop(columnas_a_eliminar, axis=1,inplace=True)
```

Desanidar columnas

```python
import ast

MoviesDataset['belongs_to_collection'] = MoviesDataset['belongs_to_collection'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])

collection_names = [] 

TotalElementos = len(MoviesDataset)

#Para iterar en los elementos y desanidar los datos:
for i in range(TotalElementos):
    try:
        if type(MoviesDataset.belongs_to_collection[i]) == dict:
            collection_names.append(MoviesDataset.belongs_to_collection[i]["name"])
        else:
            collection_names.append([])
        
    except KeyError:
        collection_names.append([])

# Crear una nueva columna con los nombres de la colección:
MoviesDataset['collection_name'] = collection_names

# Eliminar la columna anidada:
MoviesDataset = MoviesDataset.drop('belongs_to_collection', axis=1)
```

```python

MoviesDataset['genres'] = MoviesDataset['genres'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])


genre_names = [] 

TotalElementos = len(MoviesDataset)

#Para obtener los valores anidados de la columna production_companies:
for i in range(TotalElementos):
    list_genres = []
    try:
        if type(MoviesDataset.genres[i]) == list:    
            for k in range(len(MoviesDataset.genres[i])):
                list_genres.append(MoviesDataset.genres[i][k]["name"])
            genre_names.append(list_genres)
        elif type(MoviesDataset.genres[i]) == dict:
            genre_names.append(MoviesDataset.genres[i]["name"])
        else:
            genre_names.append([])
    except KeyError:
        genre_names.append([])


MoviesDataset['genre_names'] = genre_names


MoviesDataset = MoviesDataset.drop('genres', axis=1)
```

```python
#Repetimos el procedimiento en la columna production_countries:

MoviesDataset['production_countries'] = MoviesDataset['production_countries'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])


countries_names = [] 

TotalElementos = len(MoviesDataset)

for i in range(TotalElementos):
    list_countries = []
    try:
        if type(MoviesDataset.production_countries[i]) == list:    
            for k in range(len(MoviesDataset.production_countries[i])):
                list_countries.append(MoviesDataset.production_countries[i][k]["name"])
            countries_names.append(list_countries)
        elif type(MoviesDataset.production_countries[i]) == dict:
            countries_names.append(MoviesDataset.production_countries[i]["name"])
        else:
            countries_names.append([])
    except KeyError:
        countries_names.append([])


MoviesDataset['countries_names'] = countries_names

MoviesDataset = MoviesDataset.drop('production_countries', axis=1)
```

```python
#Para trasformar los valores de id del dataframe a enteros:

for i in range(TotalElementos):
    try:
        MoviesDataset.id[i] = int(MoviesDataset.id[i])
    except KeyError:
        MoviesDataset.id[i] = None
    except ValueError:
        MoviesDataset.id[i] = None        
```
